```python
from sklearn.svm import SVR
```

```python
svr = SVR()
svr.get_params()
```

```python
C = [0.001,0.01,0,1,1.0,10,100,1000]
#degree = [2,3,4,5]
kernel = ['linear','rbf','poly','sigmoid']
gamma = ['scale','auto',0.001,0.01,0.1,1.0,10.0]
#epsilon = [0.001,0.01,0,1,1.0,10,100,1000]

params = dict(C=C,kernel = kernel,gamma=gamma)

grid_svr = GridSearchCV(estimator=svr,param_grid=params,verbose=2,cv=5,scoring='neg_mean_squared_error')
grid_svr
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('salary',axis=1),df['salary'],test_size=0.3)
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
grid_svr.fit(x_train,y_train)
```
